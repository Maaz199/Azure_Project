```python
import pandas as pd
from sklearn import preprocessing

loan_train = pd.read_csv("train.csv")
loan_train.head(4)
```

```python
loan_train.isnull().sum()
```

```python
loan_train.shape
```

```python
loan_train['Grade'] = loan_train['Grade'].map({'A': 1, 'B': 2, 'C':3, 'D':4, 'E':5, 'F':6, 'G':7})
loan_train['Sub Grade'] = loan_train['Sub Grade'].map({'A1': 11, 'A2': 12, 'A3':13, 'A4':14, 'A5':15,
                                            'B1':21, 'B2':22, 'B3':23, 'B4':24, 'B5':25,
                                            'C1':31, 'C2':32, 'C3':33, 'C4':34, 'C5':35,
                                            'D1':41, 'D2':42, 'D3':43, 'D4':44, 'D5':45,
                                            'E1':51, 'E2':52, 'E3':53, 'E4':54, 'E5':55,
                                            'F1':61, 'F2':62, 'F3':61, 'F4':64, 'F5':65,
                                            'G1':71, 'G2':72, 'G3':71, 'G4':74, 'G5':75})
loan_train['Employment Duration'] = loan_train['Employment Duration'].map({'MORTGAGE':0, 'RENT':1, 'OWN':2})
loan_train['Verification Status'] = loan_train['Verification Status'].map({'Not Verified':0, 'Verified':1, 'Source Verified':2})
loan_train['Application Type'] = loan_train['Application Type'].map({'INDIVIDUAL':0, 'JOINT':1})
loan_train['Initial List Status'] = loan_train['Initial List Status'].map({'w':0, 'f':1})
loan_train['Payment Plan'] = loan_train['Payment Plan'].map({'n':0})
loan_train['Batch Enrolled'] = loan_train['Batch Enrolled'].apply(lambda x: int(x[3:]))
loan_train["Loan Title"] = loan_train["Loan Title"].apply(lambda x: x.lower())
label_encoder = preprocessing.LabelEncoder()
loan_train['Loan Title'] = label_encoder.fit_transform(loan_train['Loan Title'])
```

```python
loan_train.head(4)
```

```python
loan_train.info()
```

```python
drop_elements = ['ID', 'Accounts Delinquent', 'Payment Plan']
loan_train = loan_train.drop(drop_elements, axis = 1)
```

```python
loan_train.to_csv("Prepared_train.csv",index=False)
print("Task Complete... File is now ready to use.")
```
